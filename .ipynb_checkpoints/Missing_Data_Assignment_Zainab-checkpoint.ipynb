### Do the following steps for each column with missing values.

* Identity the column 
* How much precentage the missing value represents 
* What is the best way to handle these missing values and why
* Apply your method to the column

```python
#Identity the column

nullColumns = dataset.columns[dataset.isna().any()]

print("Columns that have null values:\n")

count=1
for nullColumn in nullColumns:
    print(count,"- ",nullColumn)
    count+=1
```

```python
#How much precentage the missing value represents

print("The precentage of the missing values in each null column:\n")
for nullColumn in nullColumns:
    missingValuePercentage = round((dataset[nullColumn].isnull().sum() / len(dataset[nullColumn]))*100,2)
    print(nullColumn+": "+str(missingValuePercentage)+"%")
```

# What is the best way to handle these missing values and why

1. For the Employee_Name column, the percentage of the null values is high and we can't guess the name of the person. Also, the name can't affect the analysis so we can drop it.


2. For the Salary column, we can fill the null values by the mean of the column records.


3. For the DOB column, we can't guess the date. The percentage of the null values is low so we can keep the column and fill the null values with a new category.


4. For the Sex column, we can't guess the sex of the person. The percentage of the null values is high but the sex feature maybe important for some records, so we can keep the column and fill the null values with a new category.


5. For the DateofTermination column, we can't guess the date and the percentage of the null values is very high. But the feature maybe important for some records, so we can keep the column and fill the null values with a new category.


6. For the ManagerID column, we can't guess the ID of the manager, and the percentage of the null values is very low. We can keep the column and fill the null values with a new category.


7. For the Absences column, we can fill the null values by the round of the median of the column records.

```python
dataset["Salary"].mean()
```

```python
dataset["Salary"].median()
```

```python
dataset["Absences"].mean()
```

```python
dataset["Absences"].median()
```

```python
#Apply your method to the column
# 1- Employee_Name column
dataset = dataset.drop(columns = ["Employee_Name"])

# 2- Salary column
dataset["Salary"] = dataset["Salary"].fillna(dataset["Salary"].mean())

# 3- DOB column
dataset["DOB"] = dataset["DOB"].fillna("ND")

# 4- Sex column
dataset["Sex"] = dataset["Sex"].fillna("ND")

# 5- DateofTermination column
dataset["DateofTermination"] = dataset["DateofTermination"].fillna("ND")

# 6- ManagerID column
dataset["ManagerID"] = dataset["ManagerID"].fillna("ND")

# 7- Absences column
dataset["Absences"] = dataset["Absences"].fillna(round(dataset["Absences"].median()))
```
