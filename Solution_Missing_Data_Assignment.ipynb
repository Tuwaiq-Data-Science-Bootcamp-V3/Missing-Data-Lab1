```python
df['ManagerID'].value_counts().plot(kind='bar')
plt.xlabel('ManagerID')
plt.ylabel('Frequency')
plt.title('Distribution of ManagerID')
plt.show()
```

```python
#this way replace missing values in ManagerID column by ManagerID depend on Manager Name
# it give me same same distribution
modeValy = lambda x: x.mode()[0]
df['ManagerID'].fillna(df.groupby('ManagerName')['ManagerID'].transform(modeValy), inplace=True)
```

```python
df['ManagerID'].value_counts().plot(kind='bar')
plt.xlabel('ManagerID')
plt.ylabel('Frequency')
plt.title('Distribution of ManagerID')
plt.show()
```
